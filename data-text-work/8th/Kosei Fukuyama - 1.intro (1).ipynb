# 自然言語処理

文章をデジタルデータとして扱えるようにし、データの構造化を行い、ラベル付けを行い、最小単位の単語分けていきます。

- デジタル化
- 構造化
- ラベル付け
- 単語分割

ここまでが、前処理となります。


これらを使って、全文検索を行えるようにしたり、クラスタリングしたり、感情判定したりしていきます。


## デジタル化

紙に書かれた状態だと扱えない。OCRや人手で入力を行い、デジタルデータにします。

## 構造化データ

- 文章ごとに番号などのIDで管理されている。（ファイル名、URLなど）
- 文章の目的が明確になっている。（タイトル、質問、回答、論文、意思表明など）
- タグ付けや更新日などが管理されている

## ラベル付け

文章ごとにラベルやタグ付けされていると、意味を表せる。後に、文章の分類を行う場合などに使えます。


# 単語分割

## 英語の場合

スペースで区切ることで、単語を抽出することができます。

これは、英語だけではなく多くの言語が単語をスペース区切ります。


## 日本語の場合

単語区切りにするのが難しい言語です。中国語や韓国語も同様に単語がスペースで区切りられていません。

他にも、「文字表現が2つ以上」あったり、「送り仮名による表現の揺れ」があり、「キャラクター数」が多く存在します。


以下の例で考えてみます。

`私の名前は寺田学です。１０年以上Twitterに使って情報発信をしています。`

```
['私', 'の', '名前', 'は', '寺田', '学', 'です', '。', '１', '０', '年', '以上', 'Twitter', 'に', '使っ', 'て', '情報', '発信', 'を', 'し', 'て', 'い', 'ます', '。']
 ```

# 単語の処理

以下のような処理を行い、単語を一定の形に変形する

- 文字表現を変換する（小文字にするなど）
- 全角・半角の文字をノーマライズ
- 原形表現に変換する

# 品詞でフィルタリング

単語の品詞を調べて、必要な品詞のみを出力することができる。

不要な単語を取り除く処理も必要となります。

以下のようなことを実施します。

- 名詞のみを取得
- 英語だと、 `a` や `the` を取り除く
- 助詞を取り除く
- 記号を削除

# その他

自然言語処理は幅が広く、様々な活用がされています。

- 感情表現としての、ポジティブ・ネガティブの判定
- 似た単語を見つける（全文検索で「もしかして」など）
- 同じ意味する単語で別の表現にする
