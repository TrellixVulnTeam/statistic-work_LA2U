# データの概要を捉える

千葉市の気象データ(Excelのデータ)の概要を捉えてみます。

- 平均値
- 中央値
- 最大値
- 最小値
- 標準偏差
- 統計情報
- 相関係数
- 散布図行列

```python
import pandas as pd
df = pd.read_excel('./data/data-chiba-2018.xlsx')
df.head()
```

```python
df.loc[:, "平均気温(℃)"].head()
```

```python
df.loc[:, "平均気温(℃)"].mean()  # 平均値
```

```python
df.loc[:, "平均気温(℃)"].median()  # 中央値
```

```python
df.loc[:, "平均気温(℃)"].max()  # 最大値
```

```python
df.loc[:, "平均気温(℃)"].min()  # 最小値
```

```python
df.loc[:, "平均気温(℃)"].std()  # 標準偏差
```

```python
df.describe()  # 統計情報
```

```python
df.corr()  # 相関係数
```

```python
%matplotlib inline
from pandas.plotting import scatter_matrix
_ = scatter_matrix(df, figsize=(15, 15))  # 散布図行列
```
