# 編集距離

編集距離(Levenshtein distance)とは、2つの単語間がどれだけ違っているかを表す指標です。

1回の文字の変更で違いを表されるものを「1」とします。この変更は、追加や削除、置き換えなどを言います。

## 編集距離の例

- pen - open 1文字追加
- pen - pan 1文字置き換え

これらは、編集距離1であると言えます。
日本語でも同じことが言い表すことができます。

## Pythonで編集距離計算

このライブラリはC言語に依存していて、whl形式のビルド済み形式のダウンロードが出来ないので、Cコンパイラが入っていない環境では利用できない可能性があります。Google Colaboratoryにも導入されていないので環境を作るのが難しい状況です。

```python
Levenshtein.distance("pen", "pan")
```

## 編集距離の応用例

編集距離は、スペルチェックや似た単語を探す場合に使われます。

Googleなどで検索をしたさいに、「もしかして」と表示される場合があります。
簡単なスペルミスを見つける際に使われます。

### もしかしての実装方法例

構成要素

- 検索対象単語（入力）
- 出力候補単語集(ソート済み単語リスト)
- 出力対象編集距離を決める(例えば2とか)
- 状態変化アルゴリズム(Levenshitein Autometa)

実装概略

- 「出力候補単語集」を作る
- 単語集をソートする
- 「検索対象単語」を状態変化アルゴリズムに渡す
- 先頭の文字から単語集に指定した編集距離以内の単語が存在するかを確認
- これら候補からも編集距離以内の文字が無いかを確認していく
